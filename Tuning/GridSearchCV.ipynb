```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
df.groupby('species').size()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop('species' , axis=1)
```

```python
y=df['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
  X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_KNN=KNeighborsClassifier(n_neighbors=11)
```

```python
model_KNN.fit(X_train , y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
y_knn=model_KNN.predict(X_test)
```

```python
accuracy_score(y_test , y_knn)  #Testing accuracy
```

```python
from sklearn.svm import SVC
```

```python
model_SVM=SVC(C=10 , gamma='auto' , kernel='linear')
```

```python
model_SVM.fit(X_train , y_train)
```

```python
y_svm=model_SVM.predict(X_test)
```

```python
accuracy_score(y_test , y_svm)
```

```python
### IMPLEMENTING GRID SEARCH CV
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model_SVM2=SVC(gamma='auto') # for gscv , just model should exist , not trained.
```

```python
classifier=GridSearchCV((model_SVM2) ,
                        {
                            'C' : [1,10,20,30],
                            'kernel' : ['rbf' , 'linear']
                        },
                        cv=5,
                        return_train_score=False
                        )
```

```python
classifier.fit(X,y) # In Cross validation(CV) , only Real x and y are passed( not the train test spit one.)
```

```python
classifier.cv_results_
```

```python
results=pd.DataFrame(classifier.cv_results_)
```

```python
results
```

```python
results[['param_C' , 'param_kernel' , 'mean_test_score']]
```

```python
print(classifier.best_params_)
```

```python
print(classifier.best_score_)
```
